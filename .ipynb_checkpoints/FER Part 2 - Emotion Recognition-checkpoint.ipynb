## Data learning

```python
from tensorflow import keras
```

```python
X_train_full = np.empty([1,1])
```

```python
if X_train_full.ndim!=3:
    X_train_full = np.expand_dims(X_train,axis=3)
```

```python
X_train_full.shape
```

```python
X_valid,X_train_new = X_train_full[:14000],X_train_full[14000:]
y_valid,y_train_new = y_train[:14000],y_train[14000:]
```

```python
from tensorflow import expand_dims
```

```python
model = keras.models.Sequential()
model.add(keras.layers.Conv2D(96,7,activation='relu',padding='same',input_shape=[48,48,1]))
model.add(keras.layers.MaxPooling2D(2))
model.add(keras.layers.BatchNormalization())
model.add(keras.layers.Conv2D(128,3,activation='relu',padding='same'))
model.add(keras.layers.Conv2D(128,3,activation='relu',padding='same'))
model.add(keras.layers.MaxPooling2D(2))
model.add(keras.layers.BatchNormalization())
model.add(keras.layers.Conv2D(256,3,activation='relu',padding='same'))
model.add(keras.layers.Conv2D(256,3,activation='relu',padding='same'))
model.add(keras.layers.MaxPooling2D(2))
model.add(keras.layers.BatchNormalization())
model.add(keras.layers.Flatten()) # fully connected layer
model.add(keras.layers.Lambda(lambda x: expand_dims(x, axis=-1)))
model.add(keras.layers.SimpleRNN(256, activation='relu'))
model.add(keras.layers.Dense(128,activation='relu'))
model.add(keras.layers.Dropout(0.5))
model.add(keras.layers.Dense(64,activation='relu'))
model.add(keras.layers.Dropout(0.5))
model.add(keras.layers.Dense(32,activation='relu'))
model.add(keras.layers.Dropout(0.5))
model.add(keras.layers.Dense(7,activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss="sparse_categorical_crossentropy",optimizer='sgd',metrics=["accuracy"])
```

```python
history = model.fit(X_train_new, y_train_new,epochs=5,validation_data=(X_valid,y_valid))
```

## Evaluation
